# Linear Regression from Zero

```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l # pip install d2l==1.0.0b0
```

```python
def synthetic_data(w,b,num): #@save, which mean it is contained in d2l
    x = torch.normal(0,1,(num,len(w)))
    y = torch.matmul(x,w)+b
    y += torch.normal(0,0.01,y.shape)
    return x, y.reshape((-1,1))

def data_iter(batch_size, features, labels): # 获取数据集中的某一个批次
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i:min(i+batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
true_w = torch.tensor([-1.,2.])
true_b = 4.2
features, labels, = synthetic_data(true_w, true_b, 1000)
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:,(0)].detach().numpy(), labels.detach().numpy(), 1)
```

```python
batch_size = 10
for x,y in data_iter(batch_size, features, labels):
    print(x,y)
    break
```
